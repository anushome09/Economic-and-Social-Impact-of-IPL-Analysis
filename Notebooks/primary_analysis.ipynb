### 4️⃣ Negative Population Impact of High-Risk Brands
Estimate the total population being negatively impacted by these high-risk brands (both pan-masala and betting apps). 

```python
revenue_data = revenue_demography.copy()
revenue_data
```

```python
revenue_data = revenue_data[['company', 'sector', 'income_group']]
revenue_data
```

```python
revenue_data.loc[:, 'income_group'] = (
    revenue_data['income_group']
    .str.replace('&', ',')
    .str.split(',')
)
revenue_data
```

```python
revenue_data.loc[:, 'income_group'] = (
    revenue_data['income_group']
    .apply(lambda x: x if isinstance(x, list) else [])
    .apply(lambda groups: [g.strip().title() for g in groups])
)
revenue_data
```

```python
demography_data = summary_demography.copy()
demography_data
```

```python
population_map = {
    "Lower": (150, 200),
    "Middle": (100, 150),
    "Upper-Middle": (30, 50)
}

def calculate_population_range(income_groups):
    min_pop, max_pop = 0, 0

    for group in income_groups:
        if group in population_map:
            lo, hi = population_map[group]
            min_pop += lo
            max_pop += hi

    if min_pop == 0 and max_pop == 0:
        return None

    return [min_pop, max_pop]
```

```python
revenue_data.loc[:, 'estimated_user_population'] = (
    revenue_data['income_group']
    .apply(calculate_population_range)
)
```

```python
revenue_data
```

```python
revenue_data['estimated_user_population']
```

```python
# Market share of each of the company 
revenue_data['market_share'] = [0.75, 0.025, 0.075, 0.07, 0.08, 0.10]

def actual_user_population(row):
    min_pop, max_pop = row['estimated_user_population']   # [130, 200] kind of values
    share = row['market_share']
    return [round(min_pop * share, 2), round(max_pop * share, 2)]

revenue_data['actual_user_population'] = revenue_data.apply(actual_user_population, axis=1)
```

```python
revenue_data
```

```python
# According the Internet, the negative impacts are:  28% and 5% for pan_masala_users and fantasy_gaming_users respectively
impact_rate = { 'Fantasy/App': 0.05, 'Pan Masala': 0.28 }

revenue_data['impact_rate'] = revenue_data['sector'].map(
    lambda x: impact_rate.get(x, 0)
)

def negative_impact_user_population(row):
    min_pop, max_pop = row['actual_user_population']
    impact_rate = row['impact_rate']

    return [round(min_pop * impact_rate, 2), round(max_pop * impact_rate, 2)]

revenue_data['negative_impact_user_population'] = (
    revenue_data.apply(negative_impact_user_population, axis=1)
)
```

```python
revenue_data
```

```python
# Total Estimated Actual User
actual_user_total_min = revenue_data['actual_user_population'].apply(lambda x: x[0]).sum()
actual_user_total_max = revenue_data['actual_user_population'].apply(lambda x: x[1]).sum()

print(f"Total estimated actual user: {actual_user_total_min:.1f}–{actual_user_total_max:.1f}")
```

```python
# Total Negatively impacted population
negatively_impacted_total_min = revenue_data['negative_impact_user_population'].apply(lambda x: x[0]).sum()
negatively_impacted_total_max = revenue_data['negative_impact_user_population'].apply(lambda x: x[1]).sum()

print(f"Total negatively impacted population: {negatively_impacted_total_min:.1f}–{negatively_impacted_total_max:.1f}") 
```

```python
min_percentage = negatively_impacted_total_min / actual_user_total_min * 100 
max_percentage = negatively_impacted_total_max / actual_user_total_max * 100

print(f"Percentage of negatively impacted users: {min_percentage:.1f}–{max_percentage:.1f}%")  
```
