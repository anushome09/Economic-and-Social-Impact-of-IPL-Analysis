### Data Cleaning: IPL Advertisers

```python
advertiser_data = pd.read_excel("..\\Dataset\\fact_ipl_advertisers.xlsx")
advertiser_data
```

```python
advertiser_data.drop(columns=['sno'], axis=1, inplace=True)
advertiser_data.columns
```

```python
advertiser_data['celebrity_influence'].unique()
```

```python
# Replacing the 'Very High' with 'Extremely High'
advertiser_data['celebrity_influence'] = advertiser_data['celebrity_influence'].replace('Very  High', 'Extremely High')
advertiser_data['celebrity_influence'].unique()
```

```python
# Replacing the 'Medium' with 'Mederate'
advertiser_data['celebrity_influence'] = advertiser_data['celebrity_influence'].replace('Medium', 'Moderate')
advertiser_data['celebrity_influence'].unique()
```

```python
advertiser_data['health_social_risk'].unique()
```

```python
# Spliting the info into list
advertiser_data[['health_social_risk', 'health_social_risk_cause']] = advertiser_data['health_social_risk'].str.split('-', expand=True)
advertiser_data
```

```python
# Shifting the values to null value
mask = advertiser_data['health_social_risk_cause'].isna()

advertiser_data.loc[mask, 'health_social_risk_cause'] = advertiser_data.loc[mask, 'health_social_risk']
advertiser_data.loc[mask, 'health_social_risk'] = None
```

```python
advertiser_data
```

```python
rank_mapping = {
    "Carcinogenic (Surrogate Ads)": "Extremely High",
    "Promotes Gambling": "Extremely High",
    "Moderate  (Sugar Content)": "Moderate",
    "Low ": "Low"
}

def fill_based_on_cause(row):
    if pd.isna(row['health_social_risk']):
        cause = row['health_social_risk_cause']      
        return rank_mapping.get(cause, "Not Applicable")
    return row['health_social_risk']

advertiser_data['health_social_risk'] = advertiser_data.apply(fill_based_on_cause, axis=1)
```

```python
advertiser_data
```

```python
advertiser_data['health_social_risk'].unique()
```

```python
advertiser_data['health_social_risk'] = advertiser_data['health_social_risk'].str.strip()
advertiser_data['health_social_risk'].unique()
```

```python
advertiser_data['health_social_risk_cause'].unique()
```

```python
advertiser_data['health_social_risk_cause'] = advertiser_data['health_social_risk_cause'].str.strip()
advertiser_data['health_social_risk_cause'].unique()
```

```python
advertiser_data.to_csv("..\\Dataset\\cleaned_ipl_advertisers.csv", index=False)
```

### Data Cleaning: IPL Central Contract

```python
central_contracts = pd.read_excel("..\\Dataset\\fact_ipl_central_contracts.xlsx")
central_contracts
```

```python
central_contracts.info()
```

```python
central_contracts[['contract_duration_(in_Year)', 'contract_duration_range']] = central_contracts['contract_duration'].str.split('Years', expand=True)
central_contracts
```

```python
central_contracts.info()
```

```python
# Converting the "Year" into integer
central_contracts['contract_duration_(in_Year)'] = central_contracts['contract_duration_(in_Year)'].astype(int)
central_contracts.info()
```

```python
# Removing the 'contract_duration'
central_contracts.drop(columns=['contract_duration'], axis=1, inplace=True)
central_contracts
```

**Filling the NULL values through research and external soruces**

```python
central_contracts[central_contracts['amount_in_crores_2025'].isnull()]
```

For `partner_sponsor_name`: 'CEAT'
* The deal value for this partnership is reported as **240 crore (total over a five-year term)** with the Board of Control for Cricket in India (BCCI).
* Since this **240 crore figure is for the full five-year cycle (202428**), the approximate annual value works out to about **48 crore per year** if averaged evenly.

```python
central_contracts.loc[central_contracts['contract_type'] == 'Official Strategic Timeout Partner', 
                      ['amount_in_crores_2025', 'total_deal_value_in_crores']] = [48, 240]
central_contracts.loc[central_contracts['contract_type'] == 'Official Strategic Timeout Partner']
```

For `partner_sponsor_name`: 'Wonder Cement'
* Wonder Cement signed a central sponsorship deal with the IPL (BCCI) worth **250 crore spanning 20242028**.
* So, for the year 2025, Wonder Cements sponsorship contribution is **approximately 50 crore**.

```python
central_contracts.loc[central_contracts['contract_type'] == 'Official Umpire Partner', 
                      ['amount_in_crores_2025', 'total_deal_value_in_crores']] = [50, 250]
central_contracts.loc[central_contracts['contract_type'] == 'Official Umpire Partner']
```

For `partner_sponsor_name`: 'Aramco'
* Aramcos IPL association (Orange Cap & Purple Cap sponsorship) is reported as an annual deal. Widely cited value: **65 crores per year**
* Since the deal is reported on a per-annum basis, assigning a total without confirmed tenure would be speculative

```python
central_contracts.loc[central_contracts['contract_type'] == 'Official Orange & Purple Cap Partner', 
                      ['amount_in_crores_2025']] = [65]
central_contracts.loc[central_contracts['contract_type'] == 'Official Orange & Purple Cap Partner']
```

```python
central_contracts
```

```python
central_contracts.to_csv("..\\Dataset\\cleaned_ipl_central_contracts.csv", index=False)
```
